```python
import os
import re
import json
```

```python


folder_path = 'downloaded_images'

# Lấy tất cả file ảnh có đuôi jpg/jpeg/png
image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

len(image_files)
```

```python
def extract_number(filename):
    # Lấy số trong tên file kiểu fb_xxx.jpg, nếu không có trả về số rất lớn để đứng cuối cùng
    match = re.search(r'fb_(\d+)\.jpg', filename)
    if match:
        return int(match.group(1))
    return float('inf')

# Sắp xếp theo số trong tên file
image_files_sorted = sorted(image_files, key=extract_number)
image_files_sorted
```

```python
image_files_sorted[-1]
```

```python

for idx, filename in enumerate(image_files_sorted, 5723):  # Bắt đầu từ 1
    ext = os.path.splitext(filename)[1].lower()  # Lấy phần mở rộng .jpg/.png
    new_name = f'fb_{idx}{ext}'
    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    print(f'Renamed: {filename} -> {new_name}')
```

```python

# Danh sách tên file theo đúng thứ tự
file_list = [
    'data1.json',
    'data2.json',
    'data3.json',
    'data4.json'
]

# List để lưu dữ liệu từng file
merged_data = []

for file_name in file_list:
    with open(file_name, 'r', encoding='utf-8') as f:
        data = json.load(f)
        merged_data.extend(data)  # Hoặc extend nếu dữ liệu là list

# Lưu vào 1 file mới
with open('merged.json', 'w', encoding='utf-8') as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=2)
```

```python
import random

A = list(range(1099, 10200))

# Lần 1: Lấy 300 phần tử ngẫu nhiên, không trùng nhau
pick1 = random.sample(A, 300)

# Lần 2: Loại bỏ pick1 khỏi A, rồi lấy tiếp 300 phần tử khác
remaining = list(set(A) - set(pick1))
pick2 = random.sample(remaining, 300)

print("Lần 1:", pick1)
print("Lần 2:", pick2)

# Kiểm tra giao giữa 2 lần
print("Số phần tử bị trùng:", len(set(pick1) & set(pick2)))  # Phải ra 0
```

```python
round_3 = [merged_data[i] for i in pick1]
```

```python
round_4 = [merged_data[i] for i in pick2]
```

```python
with open('round_4.json', 'w', encoding='utf-8') as f:
    json.dump(round_4, f, ensure_ascii=False, indent=2)
```
